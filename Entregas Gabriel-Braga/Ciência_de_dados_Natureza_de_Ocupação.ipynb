## **2. Qual o aumento, percentual, de trabalhadores MEI/Aut√¥nomos 2008 at√© 2020?**
Apresente o resultado em um gr√°fico de barras.

```python
# Convertendo a coluna 'Quantidade de Declarantes' para o tipo num√©rico
df_ss['Quantidade de Declarantes'] = pd.to_numeric(df_ss['Quantidade de Declarantes'], errors='coerce')

# Calculando o total de trabalhadores MEI/Aut√¥nomos para cada ano
df_ss_p2 = df_ss.groupby('Ano-Calend√°rio')['Quantidade de Declarantes'].sum()

# Calculando o aumento percentual em rela√ß√£o ao ano anterior
aumento_percentual = df_ss_p2.pct_change() * 100
```

## **3. Qual √© a distribui√ß√£o dos rendimentos tribut√°veis para diferentes categorias ocupacionais em 2020 e qual categoria apresentou a maior m√©dia de rendimentos tribut√°veis?**

```python
# Removendo caracteres n√£o num√©ricos da coluna 'Rendimentos Tribut√°veis'
df_ss['Rendimentos Tribut√°veis'] = df_ss['Rendimentos Tribut√°veis'].str.replace(r'\D', '', regex=True)

# Convertendo os valores para float, e tratando valores vazios ou nulos como NaN
df_ss['Rendimentos Tribut√°veis'] = pd.to_numeric(df_ss['Rendimentos Tribut√°veis'], errors='coerce')

# Filtrando os dados para o ano de 2020
df_2020 = df_ss[df_ss['Ano-Calend√°rio'] == 2020]

```

## **Desafio 1 - N√≠vel J√∫nior**

Crie uma proje√ß√£o para 2050 dos custos com despesas m√©dicas dos aposentados, baseando-se nos custos de 2016 at√© 2020.

```python
# Filtrando os dados para considerar apenas as despesas m√©dicas de pessoas aposentadas e os anos de 2016 at√© 2020
df_aposentados = df_ss[df_ss['Natureza de Ocupa√ß√£o'] == 'Aposentado']
despesas_medicas_aposentados_2016_2020 = df_aposentados[df_aposentados['Ano-Calend√°rio'].between(2016, 2020)].groupby('Ano-Calend√°rio')['Despesas M√©dicas'].sum()
```

```python
# Calculando a taxa de crescimento anual m√©dia das despesas m√©dicas para o per√≠odo de 2016 a 2020
taxa_crescimento_anual = despesas_medicas_aposentados_2016_2020.pct_change().mean()

# Projeto das despesas m√©dicas para 2050 com base na taxa de crescimento m√©dia
despesas_medicas_2050 = despesas_medicas_aposentados_2016_2020.iloc[-1] * (1 + taxa_crescimento_anual) ** (2050 - despesas_medicas_aposentados_2016_2020.index[-1])
```

```python
# Valor inicial (despesas m√©dicas de 2016)
valor_inicial = despesas_medicas_aposentados_2016_2020.iloc[0]

# Valor final (despesas m√©dicas de 2020)
valor_final = despesas_medicas_aposentados_2016_2020.iloc[-1]

# Calculando o crescimento percentual
crescimento_percentual = ((valor_final - valor_inicial) / valor_inicial) * 100

print("Crescimento percentual das despesas m√©dicas de 2016 at√© 2020:", crescimento_percentual, "%")
```

```python
# Plotando o gr√°fico da proje√ß√£o das despesas m√©dicas
plt.figure(figsize=(10, 6))
plt.plot(despesas_medicas_aposentados_2016_2020.index, despesas_medicas_aposentados_2016_2020.values, marker='o', linestyle='-', color='b', label='Despesas M√©dicas de Aposentados (2016-2020)')
plt.plot(2050, despesas_medicas_2050, marker='o', markersize=8, color='r', label='Proje√ß√£o para 2050')
plt.title('Proje√ß√£o das Despesas M√©dicas de Aposentados at√© 2050')
plt.xlabel('Ano')
plt.ylabel('Despesas M√©dicas')
plt.legend()
plt.grid(True)
plt.show()
```

Fa√ßa uma compara√ß√£o entre o nivel de contribui√ß√£o arrecadado de todos os tipos de Ocupa√ß√£o

```python
# Primeiro, vamos calcular a m√©dia de imposto a pagar para cada natureza de ocupa√ß√£o
media_por_ocupacao = df_ss.groupby('Natureza de Ocupa√ß√£o')['Imposto a Pagar'].mean()

# Agora, vamos agrupar os dados por ano e por natureza de ocupa√ß√£o e calcular a m√©dia de imposto a pagar
media_por_ano_ocupacao = df_ss.groupby(['Ano-Calend√°rio', 'Natureza de Ocupa√ß√£o'])['Imposto a Pagar'].mean().unstack()

```
